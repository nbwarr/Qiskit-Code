```python
from qiskit import *
```

```python
# Creating a circuit (number of qubits, number of classical bits)
circuit = QuantumCircuit(2, 2)

# Another way
#quantum_register = QuantumRegister(2) # 2 qubits
#classical_register = ClassicalRegister(2) # 2 classical bits
#circuit = QuantumCircuit(quantum_register, classical_register)

# Atil says that he prefers the first way because it is simpler

# Displays the circuit in Jupyter Notebook
circuit.draw()
```

```python
# Ensures that I can use matplotlib to draw the circuits with a better UI
%matplotlib inline
```

```python
# Better visuals when using matplotlib
circuit.draw(output='mpl')
```

```python
# Adds a hadamard gate to qubit 0
circuit.h(0)
circuit.draw(output='mpl')
```

```python
# CNOT (cx) gate
# (control qubit, target qubit)
# If qubit 0 is 1, cx applies the NOT gate to qubit 1, else nothing happens
circuit.cx(0, 1)

# Measure
# Measure qubit 0 and qubit 1, associate these measurements with classical bits 0 and 1
# Lists allow me to apply the operation to multiple objects
circuit.measure([0, 1], [0, 1])
circuit.draw(output='mpl')
```

```python
# Setting up the simulator
simulator = Aer.get_backend('qasm_simulator')

# Execution
result = execute(circuit, backend=simulator).result()
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts(circuit))
```
