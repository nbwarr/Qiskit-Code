```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
circuit = QuantumCircuit(3, 3)
circuit.x(0)
# Circuit barriers help visualize the steps
circuit.barrier()
circuit.h(1)
circuit.cx(1, 2)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
# Measure qubits 0 and 1 and put them in the classical bits 0 and 1
circuit.measure([0, 1], [0, 1])
circuit.barrier()
circuit.draw(output='mpl')
```

```python
# CNOT gate is controlled by qubit 1 and targets qubit 2
circuit.cx(1, 2)

# Control Z gate is controlled by qubit 0 and targets qubit 2
circuit.cz(0, 2)

# Measures qubit 2 and assigns it to classical bit 2
circuit.measure([2], [2])

circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
plot_histogram(result.get_counts(circuit))
```

This makes sense because the circuit controls qubit 2 (3rd qubit) to always be 1. Qiskit displays the qubit states in the order where q0 is the rightmost number and qn is the leftmost number. So no matter what states qubit 0 and qubit 1 end up as, because qubit 0 starts as 1 and the circuit teleports that information to qubit 2, qubit 2 will always be 1.
