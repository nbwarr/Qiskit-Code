```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
%matplotlib inline
```

```python
secretNumber = '111001010011'
```

```python
# Creating a quantum circuit with 1 more qubit than classical bit
# 1 qubit is an ancilla (helper) qubit so I won't be measuring it
circuit = QuantumCircuit(len(secretNumber) + 1, len(secretNumber))

# Adding a hadamard gate to the non-ancilla qubits
circuit.h(range(len(secretNumber)))

# Adding a NOT gate and a hadamard gate to the ancilla qubit
circuit.x(len(secretNumber))
circuit.h(len(secretNumber))
circuit.barrier()

# Enumerate allows me to keep track of the number of iterations with a count value that starts at 0
# [::-1]  or reversed() reverses the string
for index, one in enumerate(reversed(secretNumber)):
    print(f"index: {index} is {one}")
    
    # If the current index has a value of 1, create a CNOT gate controlled by the qubit at that
    # index, targeting the ancilla qubit
    if one == "1":
        circuit.cx(index, len(secretNumber))

# Remember secret number (current example is 1000101)
#circuit.cx(6, 7) # Far left 1
#circuit.cx(2, 7) # Middle 1
#circuit.cx(0, 7) # Far right 1
circuit.barrier()

# Adding hadamard to all non-ancilla qubits and measuring them
circuit.h(range(len(secretNumber)))
circuit.measure(range(len(secretNumber)), range(len(secretNumber)))

# This circuit is automatic
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts(circuit)
plot_histogram(counts)
```
